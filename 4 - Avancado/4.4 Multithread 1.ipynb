#### Sem Multithreading

```python
import cProfile
import time


def funcao_demorada():
    time.sleep(10)

def funcao_demorada2():
    time.sleep(20)

def main():
    for _ in range(5):
        funcao_demorada()
        funcao_demorada2()

if __name__ == "__main__":
    #iniciando o profiler
    profiler = cProfile.Profile() #Cria o obejto baseado no profiler
    profiler.enable() #Habilita o Profiler

    #inica a função principal do software
    main()
    
    profiler.disable() 
    profiler.print_stats(sort='cumulative') 
```

#### Utilizando Multihread

```python
import cProfile
import time
import threading


def funcao_demorada():
    time.sleep(10)

def funcao_demorada2():
    time.sleep(20)

def main():
    theads = []
    for _ in range(5):
        #funcao_demorada()
        #funcao_demorada2()
        thead1 = threading.Thread(target=funcao_demorada)
        thead2 = threading.Thread(target=funcao_demorada2)

        thead1.start()
        thead2.start()

        theads.append(thead1)
        theads.append(thead2)

    for thead in theads:
        thead.join()
if __name__ == "__main__":
    #iniciando o profiler
    profiler = cProfile.Profile() #Cria o obejto baseado no profiler
    profiler.enable() #Habilita o Profiler

    #inica a função principal do software
    main()
    
    profiler.disable() 
    profiler.print_stats(sort='cumulative') 
```

#### Utlizando Multiprocessing

```python
import cProfile
import multiprocessing.process
import time
import multiprocessing


def funcao_demorada():
    time.sleep(10)

def funcao_demorada2():
    time.sleep(20)

def main():
    processes = []
    for _ in range(5):
        #funcao_demorada()
        #funcao_demorada2()
        process1 = multiprocessing.process(target=funcao_demorada)
        process2 = multiprocessing.process(target=funcao_demorada2)

        process1.start()
        process2.start()

        processes.append(process1)
        processes.append(process2)

    for process in processes:
        process.join()
if __name__ == "__main__":
    #iniciando o profiler
    profiler = cProfile.Profile() #Cria o obejto baseado no profiler
    profiler.enable() #Habilita o Profiler

    #inica a função principal do software
    main()
    
    profiler.disable() 
    profiler.print_stats(sort='cumulative') 
```
