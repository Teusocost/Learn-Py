#### Grafico de linhas

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 16]

plt.plot(x, y)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Gráfico de Linha")

plt.show()
```

#### Grafico de Barras

```python
categorias = ["A", "B", "C", "D"]
valores = [10, 25, 15, 40]

plt.bar(categorias, valores,color=["green"])
plt.show()
```

#### Grafico de Pizza

```python
categorias = ["A", "B", "C", "D"]
valores = [10, 25, 15, 40]

plt.pie(
        valores, 
        labels=categorias, 
        autopct='%1.1f%%',
        startangle=90,
        colors=['red', 'blue', 'green', 'yellow']
        )
plt.title("Gráfico de Pizza")
plt.show()
```

#### Grafico de disperção

```python
import numpy as np
import matplotlib.pyplot as plt

#dados

x = np.random.randn(60)
y = np.random.randn(60)

plt.scatter(x, y, color='blue', marker='o')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão')

plt.show()
```

#### Grafico de serie multiplas

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#criar grafico:
plt.plot(x, y1, label='seno')
plt.plot(x, y2, label='cosseno')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Série Múltipla')

plt.show()
```
